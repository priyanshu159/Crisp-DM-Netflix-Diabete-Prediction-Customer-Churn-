```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/hmodi/Downloads/smartphones.csv")
```

## it is where we find null values 

```python

df.isnull().sum()
```

```python
df1=df.copy()
```

## It is where we drop null values from given columns

```python
df1.dropna(subset=['os','card','camera'],inplace=True)
```

```python
df1.isnull().sum()
```

```python
df
```

## it is where  we convert string data into numerical  data

```python
df1['price'] = (df1['price']
               .str.replace('₹', '', regex=False)
               .str.replace(',', '')
               .astype(float))
```

```python
df1
```

## it is where we fill our null value with the mean value

```python
df1['rating']=df1['rating'].fillna(df['rating'].mean())
```

```python
df1.drop_duplicates()
```

```python
df1['os'].unique()
```

```python
df2=df1.copy()
```

## It is where we remove unwanted rows from os column

```python
df2 = df2[~df2['os'].str.contains('Memory Card', na=False)].reset_index(drop=True)
df2 = df2[~df2['os'].str.contains('Bluetooth', na=False)].reset_index(drop=True)
df2 = df2[~df2['os'].str.contains('Rear Camera', na=False)].reset_index(drop=True)
df2 = df2[~df2['os'].str.contains('Browser', na=False)].reset_index(drop=True)
```

```python
df2['os'].unique()
```

```python
df2.info()
```

## It is where we use Label Encoder to Convert categorical data to numeric unique numbers 

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
df2['os']=l.fit_transform(df2['os'])
```
